```python
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# ----------------------------
# Custom Dice + BCE Loss
# ----------------------------
def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coef(y_true, y_pred)

def bce_dice_loss(y_true, y_pred):
    bce = tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)
    return bce + dice_loss(y_true, y_pred)

# ----------------------------
# Callbacks
# ----------------------------
checkpoint = ModelCheckpoint(
    "unet_best.h5",               # save best model locally
    monitor="val_loss",           # track validation loss (can also track val_dice_coef)
    save_best_only=True,
    mode="min",
    verbose=1
)

earlystop = EarlyStopping(
    monitor="val_loss",
    patience=8,                   # stop if no improvement for 8 epochs
    mode="min",
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

# ----------------------------
# Train model
# ----------------------------
EPOCHS = 50  # safe longer training

history = model.fit(
    train_gen,
    validation_data=val_gen,
    steps_per_epoch=80,
    validation_steps=20,
    epochs=EPOCHS,
    callbacks=[checkpoint, earlystop, reduce_lr]
)

# ----------------------------
# Save model for Keras & SavedModel
# ----------------------------
model.save("unet_final.h5")                 # Keras H5
model.save("unet_savedmodel")              # TensorFlow SavedModel folder

# Optional: Download in Colab
try:
    from google.colab import files
    files.download("unet_final.h5")
except:
    print("Local environment: download manually from filesystem.")
```
